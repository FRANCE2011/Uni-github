### **ANALISI VALORI NULLI (NAN)**

Il linguaggio di programmazione Python che si sta utilizzando, introduce una serie di dati mancanti:

**-NONE
-NAN**

Visualizziamo e analizziamo nello specifico i valori nulli nel dataset

funzione ***isnull*** genera una maschera booleana che indica i valori mancanti

```python
new.isnull().any  # restituisce valore booleano
# sono presenti alcuni valori nulli in determinate righe, dunque nella nostra analisi possono esserci problemi dovuti a missing value
```

##### **Del processo di esportazione non si ha informazioni, durante tale operazione possono venire a mancare alcuni dati che si sono cancellati o persi** 

E' opportuno ragionare sia sui dati mancanti autodescrittivi, che sono evidenti ed espliciti, sia sul fatto che il risultato finale dell'analisi possibilmente sarà influenzato dai dati mancanti. 

**NAN (not a number)** è il tipo di dato mancante introdotto da Python specifico per i numeri decimali, è trattato come floating point (virgola mobile), diverso da **NONE** (codifica generica per tutti i dati-object-)

Non fare un'accurata analisi dei dati mancanti è un errore dal punto di vista metodologico



***Quale conseguenze in presenza di NAN?***

è possibile procedere in due modi: 

-sostituendo in modo esplicito Nan con 0 (np.nansum fa la somma interpretando NAN come 0), oppure con altri valori specificati (data.fillna(0)); 

-eliminando il valore mancante (dropna( )).

Cancellare l'intera riga o il 50% delle righe non porta ad avere efficaci risultati, perchè si eliminerebbero dati che per una variabile sono mancanti ma per altre no. 

##### Dunque sarebbe opportuno filtrare tutte le righe che restituiscono il valore booleano FALSE in un'unica colonna. 

OPPURE:

```python
new.dropna(inplace=True) # oppure visualizzazione finale del dataset eliminando le righe in cui sono contenuti valori NaN
new
```

### Definire le informazioni grafiche

+ QUALI INFORMAZIONI VUOLE TRASMETTERE QUESTO GRAFICO A BARRE?
+ E' COMPRENSIBILE COME PRESENTAZIONE PUBBLICA?

Il grafico ***Bar Chart*** definisce la relazione tra due variabili x e y, e vuole mostrare quale degli Stati selezionati ha maggiori possibilità di usufuire dei servizi sanitari. 
Le condizioni sono le due diverse ***aree di residenza***, fattori determinanti per l'accesso ai servizi.

### Motivare la scelta della forma grafica

Da un punto di vista statistico/analitico il grafico a barre è maggiormente efficiente per rappresentare questa prima analisi, in quanto sia nel primo che nel secondo bar chart, la variabile usata come ***variabile dipendente*** (e quindi posizionata sull'asse delle y), è una variabile cardinale discreta (formata da numeri interi), ma la ***variabile*** scelta come ***indipendente*** (rappresentata dai valori della x), è una variabile categoriale o nominale, fornisce un'informazione in base alle due modalità di cui è composta (Urban e Rural) e rappresenta una stringa, per cui non era opportuno optare per un grafico che richiedesse entrambe le variabili cardinali (non è possibile convertire stringhe in float, es. funzione jointplot).

Da un punto di vista visivo, il bar chart permette di avere una visualizzazione migliore e più intuitiva. In questo caso, essendoci pochi dati, la diversa altezza delle barre favorisce la comprensione del grafico perchè mette a confronti gli Stati e le differenze sono ben distinguibili, piuttosto che avere pochi punti molto dispersi tra loro. 

### Evidenziare i benefici della comunicazione visuale
Comunicare attraverso rappresentazioni visive permette di meglio cogliere le ***informazioni***, di evidenziare le differenze o le somiglianze.
Il lettore preferirà osservare una bella tipologia di grafico colorato e ordinato piuttosto che leggere e dare la giusta interpretazione ai dati (soprattutto se statistici sono necessarie delle conoscenze più settoriali).
Una buona comunicazione visuale imprime direttamente nell'occhio di chi sta osservando i concetti chiave per comprendere i significati, i ***risultati dell'analisi*** e ciò che si vuole raccontare attraverso essa, catturandone l'attenzione visiva.
